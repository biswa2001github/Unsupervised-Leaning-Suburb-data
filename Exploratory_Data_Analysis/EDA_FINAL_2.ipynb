### 1.  Randomly Initialized SOM with Hierarchical Clustering

```python
from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data Preparation and Normalization
features = [
    'Population Density', '2012 ERP, total', '2012 ERP age 75-79, %',
    'Born overseas, %', 'Speaks LOTE at home, %', 'Unemployed, %',
    'Equivalent household income <$600/week, %', '% dwellings which are public housing',
    'Dwellings with no motor vehicle, %', 'Public Hospitals', 'Community Health Centres',
    'General Practice', 'Pharmacies', 'Primary Schools', 'Secondary Schools',
    'Residential (%)', 'Commercial (%)', 'Industrial (%)', 'Rural (%)'
]
scaler = MinMaxScaler()
data = merged_full_data[features]
data_normalized = scaler.fit_transform(data)

# Define the SOM with a Fixed Seed for Stability
som_grid_size = (10, 10)
np.random.seed(42)  # Ensure reproducibility
som = MiniSom(x=som_grid_size[0], y=som_grid_size[1], input_len=data_normalized.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(data_normalized) #initialize with randomized weights

# Convergence Parameters
max_iter = 20000  # Increased max iterations for better convergence
convergence_threshold = 1e-5  # Stricter threshold
stability_period = 20  # Increased stability period
min_epochs = 500  # Minimum epochs before checking convergence

previous_error = float('inf')
no_change_epochs = 0

# Training Loop with Convergence Check
for epoch in range(max_iter):
    som.train_random(data_normalized, 1)  # Train for one epoch
    current_error = som.quantization_error(data_normalized)  # Calculate quantization error

    # Check convergence only after minimum epochs
    if epoch >= min_epochs:
        error_change = abs(previous_error - current_error)
        print(f"Epoch {epoch + 1}, Quantization Error: {current_error}, Error Change: {error_change}")

        if error_change < convergence_threshold:
            no_change_epochs += 1
        else:
            no_change_epochs = 0

        if no_change_epochs >= stability_period:
            print(f"Convergence reached after {epoch + 1} epochs.")
            break

    previous_error = current_error
else:
    print("Reached maximum iterations without full convergence.")

# Extract SOM weight vectors for each node after convergence
weights = som.get_weights().reshape(-1, data_normalized.shape[1])  # Flatten the SOM grid to a 2D array

# Hierarchical Clustering on SOM Weights
linked = linkage(weights, method='ward')

# Dendrogram Visualization
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram on SOM Weights')
plt.xlabel('SOM Node')
plt.ylabel('Euclidean Distance')
plt.show()

# Assign Clusters and Map to Suburbs
num_clusters = 5
clusters = fcluster(linked, num_clusters, criterion='maxclust')

mapped_data = pd.DataFrame({
    'Suburb': merged_full_data['Suburb'],
    'SOM Node': [som.winner(x) for x in data_normalized],
    'Cluster': [clusters[som.winner(x)[0] * som_grid_size[1] + som.winner(x)[1]] for x in data_normalized]
})

# Display cluster assignments for each suburb
print(mapped_data.sort_values(by="Cluster"))

# Visualization
plt.figure(figsize=(10, 10))
for i, row in mapped_data.iterrows():
    w = row['SOM Node']
    plt.text(w[0] + 0.5, w[1] + 0.5, row['Suburb'],
             ha='center', va='center', fontsize=8,
             color=sns.color_palette('tab10')[row['Cluster'] - 1])

plt.title("SOM Grid with Hierarchical Clustering")
sns.heatmap(som.distance_map().T, cmap='coolwarm', square=True, cbar_kws={'label': 'Distance'}, linewidths=0.1)
plt.show()
```

### 2. With 5 clusters(fixed) ,SOM with PCA derived weights,Hierarchial Clustering,Hyperparameter tuning

```python
from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, fcluster
from hyperopt import fmin, tpe, hp, Trials
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data Preparation and Normalization
features = [
    'Population Density', '2012 ERP, total', '2012 ERP age 75-79, %',
    'Born overseas, %', 'Speaks LOTE at home, %', 'Unemployed, %',
    'Equivalent household income <$600/week, %', '% dwellings which are public housing',
    'Dwellings with no motor vehicle, %', 'Public Hospitals', 'Community Health Centres',
    'General Practice', 'Pharmacies', 'Primary Schools', 'Secondary Schools',
    'Residential (%)', 'Commercial (%)', 'Industrial (%)', 'Rural (%)'
]
scaler = MinMaxScaler()
data = merged_full_data[features]
data_normalized = scaler.fit_transform(data)

# Initialize SOM with PCA-derived weights
pca = PCA(n_components=data_normalized.shape[1])
pca_data = pca.fit_transform(data_normalized)

def som_training(params):
    learning_rate, sigma, grid_size = params
    som_grid_size = (int(grid_size), int(grid_size))

    # Initialize SOM with PCA weights
    som = MiniSom(x=som_grid_size[0], y=som_grid_size[1], input_len=data_normalized.shape[1], sigma=sigma, learning_rate=learning_rate)
    som.random_weights_init(pca_data)

    # Training with early stopping based on quantization error and silhouette score
    max_iter = 2000
    convergence_threshold = 1e-5
    stability_period = 20
    min_epochs = 200
    previous_error = float('inf')
    no_change_epochs = 0
    for epoch in range(max_iter):
        som.train_random(data_normalized, 1)  # Train for one epoch
        current_error = som.quantization_error(data_normalized)

        if epoch >= min_epochs:
            error_change = abs(previous_error - current_error)
            if error_change < convergence_threshold:
                no_change_epochs += 1
            else:
                no_change_epochs = 0
            if no_change_epochs >= stability_period:
                break
        previous_error = current_error

    # Calculate silhouette score only if there are multiple clusters
    weight_vectors = som.get_weights().reshape(-1, data_normalized.shape[1])  # Flatten the SOM grid
    linkage_matrix = linkage(weight_vectors, method='ward')
    num_clusters = 5  # Fixed number of clusters for silhouette score comparison
    clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')

    # Map each data point to its winning node and cluster
    mapped_clusters = [clusters[som.winner(x)[0] * som_grid_size[1] + som.winner(x)[1]] for x in data_normalized]
    
    # Ensure there are at least two unique clusters before calculating silhouette
    unique_clusters = len(set(mapped_clusters))
    if unique_clusters < 2:
        return 1e6  # High penalty for configurations producing only one cluster

    silhouette = silhouette_score(data_normalized, mapped_clusters)
    return -silhouette  # We aim to maximize silhouette, so we minimize its negative


# Define the search space for Hyperparameter Tuning
space = hp.pchoice(
    'hyperparameter_choice',
    [
        (1.0, (hp.uniform('learning_rate', 0.01, 0.9),
               hp.uniform('sigma', 0.1, 5.0),
               hp.quniform('grid_size', 5, 15, 1))),
    ]
)


# Objective function for Hyperparameter Optimization
def objective(params):
    learning_rate, sigma, grid_size = params
    silhouette = som_training([learning_rate, sigma, grid_size])
    return -silhouette  # We aim to maximize silhouette, so we minimize its negative

# Run Hyperparameter Optimization
trials = Trials()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=30,
            trials=trials)

# Best hyperparameters and corresponding silhouette score
best_learning_rate = best['learning_rate']
best_sigma = best['sigma']
best_grid_size = int(best['grid_size'])  # Convert grid size to integer

print(f"Best params: learning rate={best_learning_rate}, sigma={best_sigma}, grid size={best_grid_size}")

# Final SOM training with optimal hyperparameters
som_grid_size = (best_grid_size, best_grid_size)
som = MiniSom(x=som_grid_size[0], y=som_grid_size[1], input_len=data_normalized.shape[1], sigma=best_sigma, learning_rate=best_learning_rate)
som.random_weights_init(pca_data)
som.train_random(data_normalized, 1000)  # Train for final convergence

# Hierarchical Clustering on Final SOM Weights
weights = som.get_weights().reshape(-1, data_normalized.shape[1])
linked = linkage(weights, method='ward')
num_clusters = 5
clusters = fcluster(linked, num_clusters, criterion='maxclust')

mapped_data = pd.DataFrame({
    'Suburb': merged_full_data['Suburb'],
    'SOM Node': [som.winner(x) for x in data_normalized],
    'Cluster': [clusters[som.winner(x)[0] * som_grid_size[1] + som.winner(x)[1]] for x in data_normalized]
})

# Display final cluster assignments for each suburb
print(mapped_data.sort_values(by="Cluster"))


```

```python
from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, fcluster
from hyperopt import fmin, tpe, hp, Trials
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data Preparation and Normalization
features = [
    'Population Density', '2012 ERP, total', '2012 ERP age 75-79, %',
    'Born overseas, %', 'Speaks LOTE at home, %', 'Unemployed, %',
    'Equivalent household income <$600/week, %', '% dwellings which are public housing',
    'Dwellings with no motor vehicle, %', 'Public Hospitals', 'Community Health Centres',
    'General Practice', 'Pharmacies', 'Primary Schools', 'Secondary Schools',
    'Residential (%)', 'Commercial (%)', 'Industrial (%)', 'Rural (%)'
]
scaler = MinMaxScaler()
data = merged_full_data[features]
data_normalized = scaler.fit_transform(data)

#  Initialize SOM with PCA-derived weights
pca = PCA(n_components=data_normalized.shape[1])
pca_data = pca.fit_transform(data_normalized)

def som_training_with_optimal_clusters(params):
    learning_rate, sigma, grid_size = params
    som_grid_size = (int(grid_size), int(grid_size))

    # Initialize SOM with PCA weights
    som = MiniSom(x=som_grid_size[0], y=som_grid_size[1], input_len=data_normalized.shape[1], sigma=sigma, learning_rate=learning_rate)
    som.random_weights_init(pca_data)

    # Training with early stopping based on quantization error
    max_iter = 2000
    convergence_threshold = 1e-5
    stability_period = 20
    min_epochs = 200
    previous_error = float('inf')
    no_change_epochs = 0
    for epoch in range(max_iter):
        som.train_random(data_normalized, 1)  # Train for one epoch
        current_error = som.quantization_error(data_normalized)

        if epoch >= min_epochs:
            error_change = abs(previous_error - current_error)
            if error_change < convergence_threshold:
                no_change_epochs += 1
            else:
                no_change_epochs = 0
            if no_change_epochs >= stability_period:
                break
        previous_error = current_error

    # Map each data point to its winning node
    mapped_nodes = np.array([som.winner(x) for x in data_normalized])
    node_positions = np.unique(mapped_nodes, axis=0)
    flattened_nodes = {tuple(pos): i for i, pos in enumerate(node_positions)}
    mapped_data_to_nodes = np.array([flattened_nodes[tuple(node)] for node in mapped_nodes])

    # Determine optimal number of clusters using silhouette analysis
    silhouette_scores = []
    linkage_matrix = linkage(node_positions, method='ward')
    for num_clusters in range(2, 11):  # Adjust the range as needed
        clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
        mapped_clusters = clusters[mapped_data_to_nodes]  # Map clusters to original data points
        
        # Ensure there are at least two unique clusters before calculating silhouette
        if len(set(mapped_clusters)) < 2:
            return 1e6  # High penalty for configurations producing only one cluster
        
        silhouette_scores.append(silhouette_score(data_normalized, mapped_clusters))

    optimal_num_clusters = np.argmax(silhouette_scores) + 2  # Add 2 because range starts at 2

    # Final clustering with optimal number of clusters
    clusters = fcluster(linkage_matrix, optimal_num_clusters, criterion='maxclust')
    mapped_clusters = clusters[mapped_data_to_nodes]

    return -silhouette_score(data_normalized, mapped_clusters)  # We aim to maximize silhouette, so we minimize its negative

# Define the search space for Hyperparameter Tuning
space = hp.pchoice('hyperparameter_choice',
                   [
                       (1.0, (hp.uniform('learning_rate', 0.01, 0.9),
                              hp.uniform('sigma', 0.1, 5.0),
                              hp.quniform('grid_size', 5, 15, 1)))
                   ])

# Objective function for Hyperparameter Optimization
def objective(params):
    learning_rate, sigma, grid_size = params
    silhouette = som_training_with_optimal_clusters([learning_rate, sigma, grid_size])
    return silhouette  # We aim to maximize silhouette, so we minimize its negative


# Run Hyperparameter Optimization
trials = Trials()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=30,
            trials=trials)

# Extract best hyperparameters directly from the dictionary
best_learning_rate = best['learning_rate']
best_sigma = best['sigma']
best_grid_size = int(best['grid_size'])  # Convert grid size to integer

print(f"Best params: learning rate={best_learning_rate}, sigma={best_sigma}, grid size={best_grid_size}")


# Final SOM training with optimal hyperparameters
som_grid_size = (best_grid_size, best_grid_size)
som = MiniSom(x=som_grid_size[0], y=som_grid_size[1], input_len=data_normalized.shape[1], sigma=best_sigma, learning_rate=best_learning_rate)
som.random_weights_init(pca_data)
som.train_random(data_normalized, 1000)  # Train for final convergence

# Hierarchical Clustering on Final SOM Weights with Optimal Cluster Count
mapped_nodes = np.array([som.winner(x) for x in data_normalized])
node_positions = np.unique(mapped_nodes, axis=0)
flattened_nodes = {tuple(pos): i for i, pos in enumerate(node_positions)}
mapped_data_to_nodes = np.array([flattened_nodes[tuple(node)] for node in mapped_nodes])

# Determine optimal number of clusters again using silhouette analysis
silhouette_scores = []
linkage_matrix = linkage(node_positions, method='ward')
for num_clusters in range(2, 11):  # Adjust the range as needed
    clusters = fcluster(linkage_matrix, num_clusters, criterion='maxclust')
    mapped_clusters = clusters[mapped_data_to_nodes]  # Map clusters to original data points

    # Ensure there are at least two unique clusters before calculating silhouette
    if len(set(mapped_clusters)) < 2:
        continue  # Skip this number of clusters

    silhouette_scores.append(silhouette_score(data_normalized, mapped_clusters))

optimal_num_clusters = np.argmax(silhouette_scores) + 2
print("Optimal number of clusters:", optimal_num_clusters)

# Final mapping with optimal clusters
clusters = fcluster(linkage_matrix, optimal_num_clusters, criterion='maxclust')
mapped_clusters = clusters[mapped_data_to_nodes]

mapped_data = pd.DataFrame({
    'Suburb': merged_full_data['Suburb'],
    'SOM Node': [som.winner(x) for x in data_normalized],
    'Cluster': mapped_clusters
})

# Display final cluster assignments for each suburb
print(mapped_data.sort_values(by="Cluster"))

# Visualization
plt.figure(figsize=(10, 10))

# Group suburbs by their SOM node for better control over placement
grouped_data = mapped_data.groupby('SOM Node')['Suburb'].apply(list)

for node, suburbs in grouped_data.items():
    x, y = node
    # Adjust the vertical position of each suburb name within the node box
    for i, suburb in enumerate(suburbs):
        plt.text(x + 0.5, y + 0.5 + (i * 0.08), suburb,  # Adjust `0.15` to control vertical spacing
                 ha='center', va='center', fontsize=8,color='white')
                 #color=sns.color_palette("tab10")[grouped_data.index.get_loc(node) % 10])

plt.title("SOM Grid with Hierarchical Clustering (Optimized)")
sns.heatmap(som.distance_map().T, cmap='jet', square=True, cbar_kws={'label': 'Distance'}, linewidths=0.1)
plt.show()
```
