### Latent Profile Analysis (LPA) for Suburb Typologies

**Objective: Develop distinct profiles of suburb types based on socio-economic and demographic factors.**

**Approach: Use LPA to group suburbs based on patterns in Equivalent Household Income <$600/week, Speaks LOTE at Home %, Born Overseas %, and Public Housing %.**

**Insights: This technique allows to create a typology of suburbs, such as "affluent but culturally diverse," "high-density immigrant communities," or "low-income with limited services," enabling targeted policy approaches for each typology.**

```python
from sklearn.preprocessing import StandardScaler

lpa_data = merged_full_data[['Equivalent household income <$600/week, %', 
                             'Speaks LOTE at home, %', 
                             'Born overseas, %', 
                             '% dwellings which are public housing']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(lpa_data)
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(scaled_data)
```

```python
# Fit GMMs with different numbers of components
from sklearn.mixture import GaussianMixture
import numpy as np

n_components = np.arange(1, 10)
models = [GaussianMixture(n, covariance_type='full', random_state=42).fit(imputed_data) for n in n_components]

# Calculate BIC scores and select the model with the lowest score
bic_scores = [model.bic(imputed_data) for model in models]
best_n_components = n_components[np.argmin(bic_scores)]
best_gmm = GaussianMixture(best_n_components, covariance_type='full', random_state=42).fit(imputed_data)
labels = best_gmm.predict(imputed_data)
```

###  Socio-Economic Index Creation using PCA

**Objective: Create a composite index that ranks suburbs by socio-economic disadvantage.**

**Approach: Use Principal Component Analysis (PCA) on features such as Unemployment %, Equivalent Household Income <$600/week, and Speaks LOTE at Home % to create a single index.**

**Insights: This index can help compare socio-economic disadvantage across suburbs, allowing for a simplified overview of which areas may need targeted support or resources.**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Select the features to include in the index
features_for_index = merged_full_data[['Unemployed, %', 'Equivalent household income <$600/week, %', 'Speaks LOTE at home, %']]

# Standardize the features
scaler = StandardScaler()
standardized_data = scaler.fit_transform(features_for_index)
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(standardized_data)
```
